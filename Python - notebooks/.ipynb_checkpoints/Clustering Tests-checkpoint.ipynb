```python
import pandas as pd
import numpy as np

from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from tslearn.clustering import TimeSeriesKMeans
```

### Reading the data

```python
data = pd.read_csv('GlucoCheck/Data/consolidatedDataForPackage.csv')
meta = pd.read_csv('GlucoCheck/Data/consolidatedMetadata.csv')
meta = meta[meta.status != 'NAN']
data = data.loc[data['subjectId'].isin(meta.ID)]
```

```python
tsdata = []
for subjectId, d in data.groupby('subjectId'):
    l = []
    l.append(subjectId)

    for gv in d.GlucoseValue:
        l.append(gv)
    tsdata.append(l[:580])

tsdf = pd.DataFrame.from_records(tsdata)


tsdf = pd.merge(tsdf,meta,left_on=0,right_on='ID')
del tsdf['ID']


```

### Splitting the subjects into training and test such that our training set is representative of all teh classes


```python
all_subjects = data.subjectId.unique()

all_subjects = ['2133-039','1636-70-1010','1636-69-069','ID01','ID02','ID03','ID11','ID12','ID13','1636-69-001','1636-69-091','2133-004','2133-018',
                 'ID21', 'ID22','ID23','1636-69-026', '1636-69-032','1636-69-090','1636-69-114','1636-70-1005','2133-015','2133-017',
                 'ID29','ID30','ID31','1636-69-028','1636-69-035','1636-69-048','1636-69-053','1636-69-060','1636-69-064','1636-69-100'
                ]


testSubjects = ['2133-039','1636-70-1010','1636-69-069']



trainSubjects = ['ID01','ID02','ID03','ID11','ID12','ID13','1636-69-001','1636-69-091','2133-004','2133-018',
                 'ID21', 'ID22','ID23','1636-69-026', '1636-69-032','1636-69-090','1636-69-114','1636-70-1005','2133-015','2133-017',
                 'ID29','ID30','ID31','1636-69-028','1636-69-035','1636-69-048','1636-69-053','1636-69-060','1636-69-064','1636-69-100'
                ]

```

```python
tsdf = tsdf.loc[tsdf[0].isin(all_subjects)]
tsdf = tsdf.set_index([0],drop=True)
```

```python
trainingSet = tsdf.loc[tsdf.index.isin(trainSubjects)]
testingSet = tsdf.loc[tsdf.index.isin(testSubjects)]
```

```python
# trainingSet.to_csv('~/Desktop/ClusteringTraining.csv')
trainingSet
```

```python
# testingSet.to_csv("~/Desktop/ClusteringTest.csv")
testingSet
```

### Splitting data into X and y

```python
X_train = trainingSet.drop(['status'], axis=1)
y_train = trainingSet['status']

X_test = testingSet.drop(['status'], axis=1)
y_test = testingSet['status']
```

### Initializing a decision tree classifier and fitting it onto the training set

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train,y_train)
# 
```

#### Predicting the classes for our test set

```python
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

### Desiging and decision boundary plot

```python
X = tsdf.drop(['status'], axis=1)
y = tsdf['status']
y = y.map({'diabetic': 0,'pre-diabetic': 1, 'non-diabetic': 2})
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

Running PCA for dimensionality reduction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
reduced_data = pca.fit_transform(X)
```

```python
# reduced_data
```

```python
clf.fit(reduced_data,y)
```

```python
h=.02
```

```python
x_min, x_max = reduced_data[:, 0].min() - 1, reduced_data[:, 0].max() + 1
y_min, y_max = reduced_data[:, 1].min() - 1, reduced_data[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
```

```python
# Z
```

Displaying Boundary Plot

```python
# Plotting
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=y, alpha=0.8)
plt.xlabel("Feature-1",fontsize=15)
plt.ylabel("Feature-2",fontsize=15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```
