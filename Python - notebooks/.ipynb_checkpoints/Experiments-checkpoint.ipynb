```python

# univariate bidirectional lstm example
# from numpy import array
# from keras.models import Sequential
# from keras.layers import LSTM
# from keras.layers import Dense
from keras.layers import Bidirectional

import pandas as pd
import numpy as np


# univariate cnn lstm example
# from numpy import array
# from keras.models import Sequential
# from keras.layers import LSTM
# from keras.layers import Dense
# from keras.layers import Flatten
# from keras.layers import TimeDistributed
# from keras.layers.convolutional import Conv1D
# from keras.layers.convolutional import MaxPooling1D

# univariate convlstm example
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import ConvLSTM2D
```

```python
# split a univariate sequence
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
 
```

```python
df = pd.read_csv('~/Desktop/NCSA_genomics/Python - notebooks/GlucoCheck/Data/consolidatedDataForPaper.csv')     
# df['Display Time'] = df['Display Time'].apply(lambda x: pd.datetime.strptime(str(x), '%m/%d/%y %H:%M'))
df2 = pd.read_csv('~/Desktop/NCSA_genomics/Python - notebooks/GlucoCheck/Data/consolidatedDataForPackage.csv')
test_df = df2[df2['subjectId']=='1636-69-032']
```

```python
df = df.append(test_df[:1750])
test_df = test_df[1750:]
```

```python
vals = test_df.GlucoseValue
vals 
```

```python
test_df
```

```python
# define input sequence
raw_seq = df.GlucoseValue.tolist()
# choose a number of time steps
n_steps = 4
# split into samples
X, y = split_sequence(raw_seq, n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, rows, columns, features]
n_features = 1
n_seq = 2
n_steps = 2
X = X.reshape((X.shape[0], n_seq, 1, n_steps, n_features))
```

```python
# define model

model = Sequential()
model.add(ConvLSTM2D(filters=64, kernel_size=(1,2), activation='relu', input_shape=(n_seq, 1, n_steps, n_features), 
                     go_backwards=True))
model.add(Flatten())
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=3, verbose=1)
```

```python
# demonstrate prediction

preds = []
N = 4
x_input = array(vals[-N:].tolist())

k=0
while k<30:
    x_input = x_input.reshape((1, n_seq, 1, n_steps, n_features))
#     x_input = x_input.reshape((1, n_seq, n_steps, n_features))
    yhat = model.predict(x_input, verbose=0)
    x = round(yhat[0][0])
    preds.append(x)
    x_input = np.append(x_input,x)
    x_input = x_input[-N:]
    k+=1
preds
```

```python
def index_agreement(o,s):
        
    ia = 1 -(np.sum((o-s)**2))/(np.sum((np.abs(s-np.mean(o))+np.abs(o-np.mean(o)))**2))

    return ia


def createGap(df,start,end):
    df = df.reset_index(drop=True)
    #df = readData()
    l = len(df.index)
    if end>l:
        end = l
        
    for i in range(start,end):
        df['GlucoseValue'][i]=0
    
    
    return df

def detectGap(testing_data):

        l = []
        k = -1
        for i in testing_data['GlucoseValue']:
            k+=1
            if i==0:
                l.append(k)
        b = min(l)
        e = max(l)
        #print(b,e)
        gap=e-b
        # print("Gap detected!")
        return b,e,gap
```
