```python
import pandas as pd
from tslearn.clustering import TimeSeriesKMeans
```

```python
data = pd.read_csv('GlucoCheck/Data/Hall/Hall_data.csv')
data
```

```python
len(data.subjectId.unique())
```

```python
tsdata = []
for subjectId, d in data.groupby('subjectId'):
    l = []
    l.append(subjectId)

    for gv in d.GlucoseValue:
        l.append(gv)
    tsdata.append(l)

    # tsdata
```

```python
meta = pd.read_csv('GlucoCheck/Data/Hall/Hall_meta.csv')
# meta['status']
```

```python
tsdf = pd.DataFrame.from_records(tsdata)
tsdf = tsdf.set_index([0],drop=True)
tsdf['Y'] = meta['status'].tolist()
```

```python
tsdf['Y'] = tsdf['Y'].map({'diabetic': 0, 'pre-diabetic': 1, 'non-diabetic': 2})
tsdf
```

```python
X = tsdf.drop(['Y'], axis=1)
y = tsdf['Y']
```

```python
km = TimeSeriesKMeans(n_clusters=3, max_iter=50, metric="dtw", random_state=0)
```

```python
km.fit(X,y);
```

```python
km.cluster_centers_.shape
```

```python
X_test = X[:12]
X_test
```

```python
y_test = y[:12]
y_test
```

```python
y_pred = km.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
