```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
raw = pd.read_csv("~/Desktop/data_with_missing.csv")
imp = pd.read_csv("~/Desktop/imputed_data.csv")
ori = pd.read_csv("~/Desktop/original.csv")
raw = raw.drop(['Unnamed: 0'], axis=1)
imp = imp.drop(['Unnamed: 0'], axis=1)
ori = ori.drop(['Unnamed: 0'], axis=1)

raw = raw[1100:1200]
ori = ori[1100:1200]
imp = imp[1100:1200]
```

```python
raw = raw.astype({'GlucoseValue':int})
raw['Display Time'] = pd.to_datetime(raw['Display Time'])
plt.figure(figsize=(14,8))
sns.set(style="white")
fig = sns.lineplot(x = raw['Display Time'], y = raw['GlucoseValue'],data=raw, palette="tab10", linewidth=1.25)
sns.despine()
fig.set_xticklabels(labels=raw['Display Time'], rotation=90, ha='right', weight='bold', fontsize=13)
fig.set_yticklabels(labels=raw['GlucoseValue'], weight='bold', fontsize=13)
fig.set_xlabel('Timestamp', weight='bold', fontsize=16)
fig.set_ylabel('Glucose Value', weight='bold', fontsize=16)
```
