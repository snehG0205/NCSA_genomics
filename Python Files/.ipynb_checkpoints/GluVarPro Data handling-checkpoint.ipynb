```python
import pandas as pd
```

```python
datagvp1 = pd.read_csv('~/Desktop/NCSA_genomics/datagvp1.csv')
datagvp1
```

```python
datagvp1['Display Time'] = datagvp1['date'] +" "+ datagvp1['time']
datagvp1['Display Time'] = datagvp1['Display Time'].apply(lambda x: pd.datetime.strptime(x, '%m/%d/%y %H:%M:%S'))
datagvp1['subjectId'] = 'GVP01'
datagvp1.drop(['date','time'], axis=1, inplace=True)
datagvp1 = datagvp1.set_index(['Display Time'], drop=True)
datagvp1 = datagvp1.rename(columns={'glucose': 'GlucoseValue'})
datagvp1 = datagvp1[['subjectId', 'GlucoseValue']]
```

```python
datagvp1
```

```python
datagvp1.to_csv('~/Desktop/NCSA_genomics/Python - notebooks/TSForecasting/Data/Gluvarpro/datagvp1.csv')
```
