```python
import pandas as pd
import glob
```

```python
l = glob.glob("/Users/snehgajiwala/Desktop/NCSA_genomics/Data/Ohio-Data/OhioT1DM-training/CSV/*.csv")
OhioT1DM = pd.DataFrame()
for s in l:
    data = pd.read_csv(s)
    x = s.rfind("/")
    y = s.rfind('-')
    ID = s[x+1:y]
    y = ID.find('-')
    ID = "OD"+ID[0:y]
    data['subjectId'] = ID
    data = data.rename(columns={'_value': 'GlucoseValue','_ts': 'Display Time'})
    data['Display Time'] = data['Display Time'].apply(lambda x: pd.datetime.strptime(x, '%d-%m-%Y %H:%M:%S'))
    data = data.set_index(['Display Time'], drop=True)
    data = data[['subjectId', 'GlucoseValue']]
    OhioT1DM = OhioT1DM.append(data)
display(OhioT1DM)
OhioT1DM.to_csv('/Users/snehgajiwala/Desktop/NCSA_genomics/Data/Ohio-Data/OhioT1DM-training/OhioT1DM.csv')
```
